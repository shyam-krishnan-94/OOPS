```python
#Generators

#Helps in efficient memory management

def square(n):
  for i in range(3):
    return i**2

square(3)
#As you can see here, when I call this function, I only get 0 which means after executing the first pass, the function returns
#Ideally, I would like the answer as 4 (0,1,4), To do that, we use something called the yield keyword
```

```python
#Generators

#Helps in efficient memory management

def square(n):
  for i in range(3):
    yield i**2

for i in square(3):
  print(i)

#As you can see, we use the yield keyword, so the values are stored in a 'generator' object and outside the function, we loop over the object to get the values
#We can also use the next keyword to iterate over the values. See below illustration

```

```python
a=square(3)
```

```python
try:
  print(next(a))
except:
  print("No more elements")
```

```python
def generator():
  yield 1
  yield 2
  yield 3

#As you can see, you can return multiple values with the help of yield keyword
a=generator()
for i in a:
  print(i)
```

```python
#Decorators

#sub topic: function copy
#Here I am declaring a function

def welcome():
  return "Hi"
```

```python
# I am 'copying' a function into another variable
wel=welcome
print(wel())
```

```python
#Now I am deleting the original function
del welcome
```

```python
#As you can see the copied function will still be there
print(wel())
```

```python
#The call to the original function will return an error as this has been deleted
print(welcome())
```

```python
#Closure Functions - A function that encompasses another function

#As you can see here welcome is the closure function which encompasses another function
#The variable declared inside a closure function will be available to use in the sub-function also
def welcome():
  msg="Hi"
  def sub_welcome():
    print("welcome to the course")
    print(msg)
  return sub_welcome()

```

```python
welcome()
```

```python
#Closure Functions - A function that encompasses another function

#As you can see here welcome is the closure function which encompasses another function
#The variable declared inside a closure function will be available to use in the sub-function also
def welcome(func,lst):
  msg="Hi"
  print(func(msg))
  def sub_welcome():
    print("welcome to the course")
    print(msg)
    print(func(lst))
  return sub_welcome()
```

```python
#As you can see you can pass whole functions to this to do other activities
welcome(len,[1,2,3,4])
```

```python
#Now coming to Decorators

def repeat_twice(func):
    """A decorator that calls the decorated function two times."""

    # 1. The wrapper function takes over the job of the original function.
    def wrapper():
        func()
        # ___________________________________________

        func()
        # ___________________________________________

    # 2. The decorator returns the wrapper function.
    return wrapper

@repeat_twice
def say_hello():
    print("Hello!")
#say_hello=repeat_twice(say_hello)

# --- Test ---
print("--- Calling the decorated function ---")
say_hello()
```
